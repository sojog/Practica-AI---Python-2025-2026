```python
import requests
```

```python
BASE_URL = "http://localhost:11434"
```

```python
GENERATE_PATH =  BASE_URL + "/api/generate"
```

```python
MODEL = "gemma3:270m"
```

```python
response = requests.post(GENERATE_PATH, json=payload)
response
```
