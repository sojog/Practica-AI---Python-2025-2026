```python
import requests
```

```python
BASE_URL = "http://localhost:11434"
```

! Atentie CONTEXT_PATH != GENERATE_PATH != VERSION_PATH

```python
CHAT_PATH =  BASE_URL + "/api/chat"
```

```python
MODEL = "gemma3:270m"
```

```python
PROMPT = "Explica-mi in limba romana ce este Django in Python" or input()
```

```python
MESSAGES = [{'role': 'user', 'content': 'Explica-mi in limba romana ce este Django in Python'}]
```

```python
payload = {
    "model" : MODEL,
    "messages" : MESSAGES,
    "stream" : False
}
```

```python
response = requests.post(CHAT_PATH, json=payload)
response
```

```python
response.content
```

```python
with open("chat_history.txt", "wb") as file_writer:
    file_writer.write(response.content)
```

```python
response.json()
```

```python
actual_reply_message = response.json()["message"]
actual_reply_message
```

```python
actual_reply_content = response.json()["message"]['content']
actual_reply_content
```

```python
MESSAGES.append(actual_reply_message)
MESSAGES
```
